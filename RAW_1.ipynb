### The Third error HttpError 404 indicates that the YouTube Data API could not find the video identified by the provided videoId parameter. The error message states: "The video identified by the videoId parameter could not be found."


```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import time

# Read CSV file
df = pd.read_csv("/Users/salmanmohamm3d/Desktop/Social M A/vdoLinks.csv")

def get_video_comments(video_id, max_results=100):
    youtube = build('youtube', 'v3', developerKey='AIzaSyAb3gjjY89XQ7nbEKIbRU-qB0AcrhRyLTo')
    comments = []
    next_page_token = None

    try:
        while True:
            request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=max_results,
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
                comments.append(comment)

            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break

    except HttpError as e:
        if 'commentsDisabled' in str(e):
            print(f"Comments are disabled for video ID: {video_id}. Skipping this video.")
            return []
        else:
            raise e

    return comments

video_comments = []

# Counter to track the number of processed video IDs
processed_count = 0
max_videos = 10000

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    if processed_count >= max_videos:
        break
    video_id = row['youtubeId']  # Assign video_id inside the loop
    print(f'Fetching comments for video ID: {video_id}')
    comments = get_video_comments(video_id)
    if comments:
        for comment in comments:
            video_comments.append({'video_id': video_id, 'comment': comment})
    processed_count += 1

# Convert the video_comments list to a DataFrame
comments_df = pd.DataFrame(video_comments)

# Write the DataFrame to a CSV file
comments_df.to_csv('video_comments.csv', index=False)

print('CSV file has been created successfully.')
```
